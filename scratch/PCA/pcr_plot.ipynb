```python
%matplotlib inline

import numpy as np 
from matplotlib import pyplot as plt 
import pickle
import glob
```

```python
SAVEFIG_OPTS = {
    'dpi' : 300,
    'transparent': True
}
```

```python
# Import pickled data as a list of dictionaries
data = [pickle.load(open(f, 'rb')) for f in glob.glob('PCR*.p')]
```

```python
def plot_eigenvalues(evals, title, ax=None, **kwds):    
    plt.bar(range(1,len(evals)+1), evals)
    plt.xlabel(r'$i$', fontsize=18)
    plt.ylabel(r'$\sigma_i$', fontsize=18)
    plt.title('Eigenvalues for %s' % title)
    yrange = np.max(evals)-np.min(evals)
    plt.ylim((np.min(evals)-yrange/50, np.max(evals)+yrange/50))
    plt.show()
    
#     if ax is None:
#         fig = plt.figure(figsize=(12,12))
#         ax = fig.add_subplot(111)    
        
# fig = plt.figure(figsize=(12,12))
# ax = fig.add_subplot(111)
# plot_eigenvalues()
```

```python
# Plot datasets
titles = ['Median NN', 'Polar-Azimuth Median NN', 'Polar-Azimuth Mean Volume', 
              'RD-TD-Radial Median NN', 'RD-TD-Radial Mean Volume', 'Mean Volume']
fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(111)
plot_eigenvalues(np.diag(data[0]['s']), titles[0], ax=ax)
fig.savefig('MedNN.png', **SAVEFIG_OPTS)

```
